```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
def mean_fits(files):
    size = len(files)
    
    if size > 0:
        
        hdul = fits.open(files[0])
        data = hdul[0].data
        hdul.close()
        
        for i in range(1, size):
            hdul = fits.open(files[i])
            data += hdul[0].data
            hdul.close()
        
        return data/size
    
```

```python
mean = mean_fits(['image0.fits', 'image1.fits', 'image2.fits', 'image3.fits', 'image4.fits'])

# Plot the 2D image data
plt.imshow(mean.T, cmap=plt.cm.viridis)
plt.colorbar()
plt.show()
```
