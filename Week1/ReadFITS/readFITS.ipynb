```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_fits(fit):
    """Calculated max value from file fits
    
    Arguments:
        fit {fits} -- file fits
    Returns:
        float -- max value position from fits
    """
    hdulist = fits.open(fit)
    data = hdulist[0].data
    
    arg_max = np.argmax(data)
    max_pos = np.unravel_index(arg_max, data.shape)
    return max_pos
```

```python
if __name__ == '__main__':
    img = 'image0.fits'
    bright = load_fits(img)
    print(bright)

    hdulist = fits.open(img)
    data = hdulist[0].data

    plt.imshow(data.T, cmap=plt.cm.viridis)
    plt.colorbar()
    plt.show()
```
